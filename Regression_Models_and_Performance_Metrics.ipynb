Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose?

Ans:

It is a method to understand the realtionship between one independent variable (X) and one dependent variable (Y)

Formula : Y = mX + C

Y = dependent variable

X = independent variable

m = slope

c = intercept

It is used to establish a linear relationship between X and Y

There is little or no multicollinearity

Homoscedasticity

Errors are normally distributed


Question 2: What are the key assumptions of Simple Linear Regression?

Ans:

a. Linearity : That the relationhsip between independent variable (X) and dependent variable (Y) is linear - straight line

b. Independence of Errors : The errors are independent of each other. There is no corelation between errors

c. Homoscedasticity : The variance of errors are constant across all values of X.

d. Normality of Errors : The errors are normally distributed.

e. No Multicollinearity : Since SLR has one independent and one dependent variable, this assumption is automatically satisfied.

f. No Significant outliers : Extreme values can influence the regression line

Question 4: Provide a real-world example where simple linear regression can be
applied.

Ans:

A real world example can be house price and house size. The more is the house size, the more is the price. Here house size is (X) the independent variable and price (Y) is the dependent variable

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
data = {
    'House_Size (sq ft)' : [600, 800, 1000, 1200, 1500],
    'House_Price (lakhs)' : [25, 32, 40,  48, 60]
}

df  = pd.DataFrame(data)
df.head()
```

```python

line = LinearRegression()
line.fit(df[['House_Size (sq ft)']], df['House_Price (lakhs)'])
```

```python
plt.scatter(df['House_Size (sq ft)'], df['House_Price (lakhs)'])
plt.plot(df['House_Size (sq ft)'], line.predict(df[['House_Size (sq ft)']]))
plt.xlabel('House_Size (sq ft)')
plt.ylabel('House_Price (lakhs)')
```

Question 5: What is the method of least squares in linear regression?

Ans:

The method of least squares is a technique used to find the best-fitting regression line in linear regression

It estimates the regression coefficients such that the sum of the squared differences between the actual values and the predicted values is minimum.

∑(yact​−ypred)^2

yact = actual value

ypred= predited value

The values are squared,so that the positive and negative values do notcancel out each other, it penalizes large error more heavily.

Question 7: Name and briefly describe three common evaluation metrics for regression
models.

Ans:

a. Mean Absolute Error :
Measures the average absolute difference between actual and predicted values. It is easy to understand and treats all errors equally.

MAE= 1/n summation |yact - ypred|

b. Mean Squared Error :

Measures the average of squared errors between actual and predicted values. Penalizes large error more heavily.

MSE = 1/n summation (yact - ypred)^2

c. Root Mean Squared Error :
Square root of MSE expressed in the same units as the target variable .

RMSE = root of MSE

Question 8: What is the purpose of the R-squared metric in regression analysis?

Ans:

R square also called the Coeff of Determination, it measures how well the regression model explains the variability of dependent variable.

It represents the proportion of variance in the dependent variable(Y) that is explained by independdent variable (X).

R² = 0 → Model explains none of the variability

R² = 1 → Model explains all the variability

R² = 0.75 → 75% of the variation in Y is explained by the model

Accuracy - Helps assess model goodness of fit

Allows comparison of regression models

Indicates how well predictors explain the outcome

Question 10: How do you interpret the coefficients in a simple linear regression model?

Ans:

A SLR is written as

Y = mX + c

c = represents the value when X is 0. It is the starting point of the regression line.

If c = 20, then X = 0 and Y = 20

m = indicates the slope, the change in Y for a one unit increase in X. If there is a positive slope, Y increases as X increases, if there is a negative slope, Y decreases as X decreases.

Conclusion - In SLR, the intercept shows the expected value of the dependent variable when independent variable is 0. While slope shows how much the dependent variable changes for one-unit increase in the independent variable.
