1.  What are data structures, and why are they important?

     Data Structures is way of organizing and storing data so that it can be acccessed and manipulated effeciently.
     Ex: string, list, set, tuples, dictionary, array.
     They are very Efficient, scalable and Flexibility.



2. Explain the difference between mutable and immutable data types with examples?

    Mutable data types can be changed after creation (Ex:lists), while immutable data types cannot be changed once created (Ex:strings).

3. What are the main differences between lists and tuples in Python?

    Lists are mutable and can be modified after creation, while tuples are immutable and cannot be changed once defined. Lists use square brackets[], whereas tuples use parentheses().

4. Describe how dictionaries store data?

    Dictionaries in Python store data as key-value pairs. Each key is unique and maps to a specific value.



5. Why might you use a set instead of a list in Python?

    we might use a set instead of a list in Python when you need to store unique elements and perform fast membership tests.Sets automatically remove duplicates and offer faster lookup compared to lists.
  

6.  What is a string in Python, and how is it different from a list?

    A string in Python is a sequence of characters enclosed in quotes (Ex:"hello"), and it is immutable, meaning it cannot be changed after creation. A list is a collection of items (Ex: [1, 2, 3]) that is mutable, so its elements can be modified.


7. How do tuples ensure data integrity in Python?

    Tuples ensure data integrity in Python by being immutable,meaning their contents cannot be changed after creation. This prevents accidental modification and makes them reliable for storing fixed collections of data.



8. What is a hash table, and how does it relate to dictionaries in Python?

    A hash table is a data structure that stores data in an associative manner using key-value pairs.A dictionary is an implementation of a hash table. When you create a dictionary, Python uses hashing to map each key to its corresponding value, allowing for fast lookups, insertions, and deletions.So, dictionaries in Python are built on the concept of hash tables.

9. Can lists contain different data types in Python?

    Yes, lists in Python3 can contain different data types. Ex: my_list = [1, "Mounika", 3.14, True]


10.  Explain why strings are immutable in Python?

      Strings are immutable in Python because once a string is created, its contents cannot be changed. This design ensures data integrity, security, and performance—especially when strings are used as keys in dictionaries or shared across multiple parts of a program.Instead of modifying a string, Python creates a new one when changes are needed.

  

11. What advantages do dictionaries offer over lists for certain tasks?

    Dictionaries offer faster lookups using keys and allow association of values with unique identifiers,making them ideal for tasks like storing user data,configurations,or counting items.Unlike lists,you don’t need to search through indexes to find a specific value.

12. Describe a scenario where using a tuple would be preferable over a list.

    Use a tuple when storing fixed data like coordinates(x, y) or dates(day, month, year)that shouldn't change,ensuring data integrity.

13. How do sets handle duplicate values in Python?

    Sets automatically remove duplicates,storing only unique elements,which helps in filtering and membership testing.

14. How does the “in” keyword work differently for lists and dictionaries?
     
    In lists-"in" checks for values whereas in dictionaries-it checks for keys, not values.

15. Can you modify the elements of a tuple? Explain why or why not?

    No,tuples are immutable,meaning their elements cannot be changed once defined.

16.  What is a nested dictionary, and give an example of its use case?

     A nested dictionary is a dictionary inside another dictionary.Ex:students = {'John': {'age': 20, 'grade': 'A'}}


17. Describe the time complexity of accessing elements in a dictionary?

    Accessing elements in a dictionary has O(1) time complexity due to hashing.

18. In what situations are lists preferred over dictionaries?

    Lists are preferred when order matters or when storing sequential data without key-value pairs.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?

    Dictionaries were unordered before Python 3.7,meaning item order wasn't guaranteed,affecting predictable iteration.

20. Explain the difference between a list and a dictionary in terms of data retrieval.

    Lists retrieve data by index, while dictionaries retrieve data by key, offering faster access for labeled data.
