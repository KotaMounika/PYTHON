1 : What is the difference between multithreading and multiprocessing?

   Multithreading in Python allows multiple threads to run within a single process, sharing memory space. It's ideal for I/O-bound tasks but limited by the Global Interpreter Lock (GIL).
   Multiprocessing uses separate processes with independent memory, enabling true parallelism for CPU-bound tasks and bypassing the GIL.


2: What are the challenges associated with memory management in Python?

  Python memory management challenges include memory leaks,fragmentation,and performance overhead from garbage collection.These issues can affect efficiency,especially in large or real-time applications.
  Ex:- Objects that are no longer needed but still referenced can accumulate, preventing memory from being freed. This often happens with circular references or poorly managed caches.


```python
#3:Write a Python program that logs an error message to a log file when a division by zero exception occurs.
try:
    # Division operation
    a = 10
    b = 0
    result = a / b
    print("Result:", result)

except ZeroDivisionError:
    # Log error to a file
    with open("error_log.txt", "a") as log_file:
        log_file.write("Error: Division by zero occurred.\n")
print("Program continues after handling the error.")
```

```python
#4:Write a Python program that reads from one file and writes its content to another file.

# Read from source file and write to destination file

# Open the source file in read mode
with open("source.txt", "r") as source_file:
    content = source_file.read()

# Open the destination file in write mode
with open("destination.txt", "w") as destination_file:
    destination_file.write(content)

print("Content copied successfully.")
```

```python
#5: Write a program that handles both IndexError and KeyError using a try-except block
# Program to handle IndexError and KeyError
my_list = [10, 20, 30]
my_dict = {'a': 1, 'b': 2}
try:
    # This will raise IndexError
    print("List item:", my_list[5])
    # This will raise KeyError
    print("Dictionary value:", my_dict['z'])
except IndexError:
    print("Oops! That index is out of range.")
except KeyError:
    print("Oops! That key does not exist in the dictionary.")
print("Program continues after handling exceptions.")
```

6: What are the differences between NumPy arrays and Python lists?

   NumPy arrays and Python lists differ primarily in performance, memory efficiency, and functionality. NumPy arrays are homogeneous, meaning all elements must be of the same data type, which allows for faster computation and lower memory usage. They support advanced mathematical operations and broadcasting, making them ideal for numerical and scientific tasks. In contrast, Python lists are heterogeneous and more flexible, capable of storing mixed data types, but they are slower and consume more memory. Lists require explicit loops for element-wise operations, whereas NumPy arrays can perform these operations efficiently using vectorization

```python
# 10: Calculate the correlation matrix using Seaborn and visualize it with a
# heatmap.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Math': [90, 80, 70, 60, 50],
    'Science': [85, 75, 65, 55, 45],
    'English': [88, 78, 68, 58, 48],
    'History': [92, 82, 72, 62, 52]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

# Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title("Correlation Matrix Heatmap")

# Show plot
plt.show()
```
